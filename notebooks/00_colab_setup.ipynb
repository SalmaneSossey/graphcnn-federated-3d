# 00 - Colab Setup

Run this notebook first on a Google Colab T4 runtime. It clones or updates the GitHub repo, installs Colab-safe dependencies without replacing Colab's CUDA PyTorch, verifies the GPU, and runs smoke tests.

```python
REPO_URL = "https://github.com/SalmaneSossey/graphcnn-federated-3d.git"
REPO_DIR = "/content/graphcnn-federated-3d"
BRANCH = "main"

print("Repo:", REPO_URL)
print("Directory:", REPO_DIR)
print("Branch:", BRANCH)
```

```python
from pathlib import Path
import os
import subprocess

def run(command):
    print("$", " ".join(command))
    subprocess.run(command, check=True)

repo_path = Path(REPO_DIR)
if repo_path.exists():
    os.chdir(REPO_DIR)
    run(["git", "fetch", "origin"])
    run(["git", "checkout", BRANCH])
    run(["git", "pull", "--ff-only", "origin", BRANCH])
else:
    run(["git", "clone", "--branch", BRANCH, REPO_URL, REPO_DIR])
    os.chdir(REPO_DIR)

print("Current directory:", os.getcwd())
```
